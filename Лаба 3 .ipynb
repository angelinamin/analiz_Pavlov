 ### 1.–ò–º–µ–µ—Ç—Å—è —Å–ø–∏—Å–æ–∫ –∏–∑ 20 —Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 100. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞–∑–±–∏–≤–∞–µ—Ç —ç—Ç–æ—Ç —Å–ø–∏—Å–æ–∫ –Ω–∞  ùëõ —Å–ø–∏—Å–∫–æ–≤ –ø–æ ùëò —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∏ –≤–∫–ª—é—á–∞–µ—Ç –≤—Å–µ —ç—Ç–∏ —Å–ø–∏—Å–∫–∏ –≤ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ  ùëò –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–∏.

````python
import random 

def —Ä–∞–∑–±–∏—Ç—å_—Å–ø–∏—Å–æ–∫(–∏—Å—Ö–æ–¥–Ω—ã–π_—Å–ø–∏—Å–æ–∫, k): 
   
    
    –ø–æ–¥—Å–ø–∏—Å–∫–∏ = [–∏—Å—Ö–æ–¥–Ω—ã–π_—Å–ø–∏—Å–æ–∫[i:i+k] for i in range(0, len(–∏—Å—Ö–æ–¥–Ω—ã–π_—Å–ø–∏—Å–æ–∫), k)] 
  
    
    return –ø–æ–¥—Å–ø–∏—Å–∫–∏ 
–∏—Å—Ö–æ–¥–Ω—ã–π_—Å–ø–∏—Å–æ–∫ = [random.randint(0, 100) for _ in range(20)]


k = 5  
—Ä–µ–∑—É–ª—å—Ç–∞—Ç = —Ä–∞–∑–±–∏—Ç—å_—Å–ø–∏—Å–æ–∫(–∏—Å—Ö–æ–¥–Ω—ã–π_—Å–ø–∏—Å–æ–∫, k)  

print(—Ä–µ–∑—É–ª—å—Ç–∞—Ç)  
````

2.–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –≤—ã—á–∏—Å–ª—è—é—â—É—é —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞, –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º–æ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞.

````python
def factorial(n):
    result = 1
    
    for i in range(1, n + 1):
        result *= i
        
    return result

n = int(input("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ n: "))
result = factorial(n)

print(f"–§–∞–∫—Ç–æ—Ä–∏–∞–ª —á–∏—Å–ª–∞ {n} —Ä–∞–≤–µ–Ω {result}")
````

3.–ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é —É–º–Ω–æ–∂–µ–Ω–∏—è, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ —Ç—Ä–µ—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤. –§—É–Ω–∫—Ü–∏—é –≤—ã–∑–≤–∞—Ç—å —Å –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–º–∏ –Ω–∏–∂–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏. –î–ª—è —Å–ª—É—á–∞—è a4 –≤—ã–±—Ä–∞—Ç—å 3 –ø–µ—Ä–≤—ã—Ö –∏ 3 –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è.

a1 = (15, 10 ,5)
a2 = (3, 1)
a3 = [2, 35, 55]
a4 = (5, 10 ,15, 20)

````python
def multiplication(a, b=None, c=None):
   
    result = 1  
    if a is not None:
        result *= a  
        result *= b  
    if c is not None:
        result *= c 
    return result 

a1 = (15, 10, 5)  
a2 = (3, 1)  
a3 = [2, 35, 55]  
a4 = (5, 10, 15, 20) 

result_a1 = multiplication(*a1)  
result_a2 = multiplication(*a2)  
result_a3 = multiplication(*a3)  
result_a4_first = multiplication(*a4[:3])
result_a4_last = multiplication(*a4[-3:])


print("–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è  a1:", result_a1)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è  a2:", result_a2)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è  a3:", result_a3)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ø–µ—Ä–≤—ã—Ö —Ç—Ä–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π  a4:", result_a4_first)
print("–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Ç—Ä–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π  a4:", result_a4_last)
````

4.–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Ä–∞–¥–∏—É—Å–∞ r –≤—ã—á–∏—Å–ª—è–µ—Ç –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–æ—Ä—Ç–µ–∂ –∏–∑ 2 –∑–Ω–∞—á–µ–Ω–∏–π.

````python
import math

def  calculate_circle_properties():
    #–ø–ª–æ—â–∞–¥—å
    square = math.pi * radius**2
     #–¥–ª–∏–Ω–∞
    circumference_length = 2 * math.pi * radius

    return (square, circumference_length)


result = calculate_circle_properties()

radius = float(input("–í–≤–µ–¥–∏—Ç–µ —Ä–∞–¥–∏—É—Å –∫—Ä—É–≥–∞: "))

print(f"–ü–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞: {result[0]}, –î–ª–∏–Ω–∞ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏: {result[1]}")
````

 5.–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —á–µ—Ç—ã—Ä–µ—Ö –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π, –ø—Ä–µ–æ–±—Ä–∞–∑—É—é—â–∏—Ö —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ –≤ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ. –°–æ–∑–¥–∞—Ç—å —Å–ª–æ–≤–∞—Ä—å —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∏–º —Å–∏–º–≤–æ–ª–∞–º–∏ –æ–ø–µ—Ä–∞—Ü–∏–π. –î–ª—è –¥–≤—É—Ö –∑–∞—Ä–∞–Ω–µ–µ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª (–Ω–∞–ø—Ä–∏–º–µ—Ä, 25 –∏ 4) –≤—ã–ø–æ–ª–Ω–∏—Ç—å –≤—ã–±—Ä–∞–Ω–Ω—É—é –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –æ–ø–µ—Ä–∞—Ü–∏—é.

````python
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiplication(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ"
    return x / y

#—Å–ª–æ–≤–∞—Ä—å
operations = {
    '+': add,
    '-': subtract,
    '*': multiplication,
    '/': divide
}




operation_symbol = input("–í–≤–µ–¥–∏—Ç–µ —Å–∏–º–≤–æ–ª  (+, -, *, /): ")

num1 = int(input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ: "))
num2 = int(input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ: "))

if operation_symbol in operations:
   
    result = operations[operation_symbol](num1, num2)
   
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç {num1} {operation_symbol} {num2} = {result}")
   
else:
    print("–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Å–∏–º–≤–æ–ª.")
   
````

6.–ù–∞–ø–∏—Å–∞—Ç—å –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä –¥–ª—è —Å—Ç—Ä–æ–∫–æ–≤—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π –≤–∏–¥–∞ '<—á–∏—Å–ª–æ> <–æ–ø–µ—Ä–∞—Ü–∏—è> <—á–∏—Å–ª–æ>', –≥–¥–µ <—á–∏—Å–ª–æ> - —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –Ω–∞–ø—Ä–∏–º–µ—Ä 113, <–∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è> - –æ–¥–Ω–∞ –∏–∑ –æ–ø–µ—Ä–∞—Ü–∏–π +,-,*,//,%,^. –ü—Ä–∏–º–µ—Ä: calc('13 - 5') -> 8

````python
def calc(term):

    components = term.split()  
    chislo1 = float(components[0]) 
    operation = components[1]
    chislo2 = float(components[2]) 

    
    if operation == '+':
        result = chislo1 + chislo2
    elif operation == '-':
        result = chislo1 - chislo2
    elif operation == '*':
        result = chislo1 * chislo2
    elif operation == '//':
        # –î–µ–ª–µ–Ω–∏–µ —Å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º 
        if chislo2 != 0:
            result = chislo1 // chislo2
        else:
            result = "–û—à–∏–±–∫–∞: –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å"
    elif operation == '%':
        result = chislo1 % chislo2
    elif operation == '^':
        result = chislo1 ** chislo2
    else:
        result = "–û—à–∏–±–∫–∞: –Ω–µ–ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è"

    return result

term = input("–í–≤–µ–¥–∏—Ç–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –≤ —Ñ–æ—Ä–º–∞—Ç–µ '—á–∏—Å–ª–æ1 –æ–ø–µ—Ä–∞—Ü–∏—è —á–∏—Å–ª–æ2': ")
result = calc(term)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç –≤—ã—Ä–∞–∂–µ–Ω–∏—è '{term}': {result}")

````

````python
def to_int(—Ç–µ–∫—Å—Ç):
    
    
    —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4,
        '–ø—è—Ç—å': 5, '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9,
        '–¥–µ—Å—è—Ç—å': 10, '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14,
        '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18, '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19,
        '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40, '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50,
        '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60, '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80, '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90,
        '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400, '–ø—è—Ç—å—Å–æ—Ç': 500,
        '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800, '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

   
    —Å–ª–æ–≤–∞ = —Ç–µ–∫—Å—Ç.split()

    —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ = 0
    –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç = 0

    for —Å–ª–æ–≤–æ in —Å–ª–æ–≤–∞:
        if —Å–ª–æ–≤–æ in —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ:
            —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ += —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ[—Å–ª–æ–≤–æ]  
        else:
            —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ = 0

  
    –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç += —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ

    return –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç


–≤–∞—à_—Ç–µ–∫—Å—Ç = input("–í–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (–Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ): ")
—Ä–µ–∑—É–ª—å—Ç–∞—Ç = to_int(–≤–∞—à_—Ç–µ–∫—Å—Ç)
print(f"–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∏–∑ —Ç–µ–∫—Å—Ç–∞ '{–≤–∞—à_—Ç–µ–∫—Å—Ç}': {—Ä–µ–∑—É–ª—å—Ç–∞—Ç}")
````

````python
def to_int(—Ç–µ–∫—Å—Ç):
    —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ = {
        '–Ω–æ–ª—å': 0, '–æ–¥–∏–Ω': 1, '–¥–≤–∞': 2, '—Ç—Ä–∏': 3, '—á–µ—Ç—ã—Ä–µ': 4,
        '–ø—è—Ç—å': 5, '—à–µ—Å—Ç—å': 6, '—Å–µ–º—å': 7, '–≤–æ—Å–µ–º—å': 8, '–¥–µ–≤—è—Ç—å': 9,
        '–¥–µ—Å—è—Ç—å': 10, '–æ–¥–∏–Ω–Ω–∞–¥—Ü–∞—Ç—å': 11, '–¥–≤–µ–Ω–∞–¥—Ü–∞—Ç—å': 12, '—Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å': 13, '—á–µ—Ç—ã—Ä–Ω–∞–¥—Ü–∞—Ç—å': 14,
        '–ø—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 15, '—à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç—å': 16, '—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 17, '–≤–æ—Å–µ–º–Ω–∞–¥—Ü–∞—Ç—å': 18, '–¥–µ–≤—è—Ç–Ω–∞–¥—Ü–∞—Ç—å': 19,
        '–¥–≤–∞–¥—Ü–∞—Ç—å': 20, '—Ç—Ä–∏–¥—Ü–∞—Ç—å': 30, '—Å–æ—Ä–æ–∫': 40, '–ø—è—Ç—å–¥–µ—Å—è—Ç': 50,
        '—à–µ—Å—Ç—å–¥–µ—Å—è—Ç': 60, '—Å–µ–º—å–¥–µ—Å—è—Ç': 70, '–≤–æ—Å–µ–º—å–¥–µ—Å—è—Ç': 80, '–¥–µ–≤—è–Ω–æ—Å—Ç–æ': 90,
        '—Å—Ç–æ': 100, '–¥–≤–µ—Å—Ç–∏': 200, '—Ç—Ä–∏—Å—Ç–∞': 300, '—á–µ—Ç—ã—Ä–µ—Å—Ç–∞': 400, '–ø—è—Ç—å—Å–æ—Ç': 500,
        '—à–µ—Å—Ç—å—Å–æ—Ç': 600, '—Å–µ–º—å—Å–æ—Ç': 700, '–≤–æ—Å–µ–º—å—Å–æ—Ç': 800, '–¥–µ–≤—è—Ç—å—Å–æ—Ç': 900
    }

  
    —Å–ª–æ–≤–∞ = —Ç–µ–∫—Å—Ç.split()


    —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ = 0
    –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç = 0

    for —Å–ª–æ–≤–æ in —Å–ª–æ–≤–∞:
        if —Å–ª–æ–≤–æ in —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ:
            —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ += —á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—ã–µ[—Å–ª–æ–≤–æ]
        else:
            print(f"–û—à–∏–±–∫–∞: '{—Å–ª–æ–≤–æ}' - –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ —Å–ª–æ–≤–æ –≤ —á–∏—Å–ª–µ.")
            return -1

   
    –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç += —Ç–µ–∫—É—â–µ–µ_—á–∏—Å–ª–æ

    return –æ–±—â–∏–π_—Ä–µ–∑—É–ª—å—Ç–∞—Ç


–≤–∞—à_—Ç–µ–∫—Å—Ç = input("–í–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (–Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ): ")
—Ä–µ–∑—É–ª—å—Ç–∞—Ç = to_int(–≤–∞—à_—Ç–µ–∫—Å—Ç)

if —Ä–µ–∑—É–ª—å—Ç–∞—Ç != -1:
    print(f"–ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∏–∑ —Ç–µ–∫—Å—Ç–∞ '{–≤–∞—à_—Ç–µ–∫—Å—Ç}': {—Ä–µ–∑—É–ª—å—Ç–∞—Ç}")
else:
    print("–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞. –ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –≤–≤–æ–¥.")
````
